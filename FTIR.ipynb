```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
config = {'axes.spines.right': False,
    'axes.spines.top': False,
    'axes.edgecolor': '.4',
    'axes.labelcolor': '.0',
    'axes.titlesize': 'large',
    'axes.labelsize': 'medium',
    'figure.autolayout': True,
    'figure.figsize': (4.5, 3.5),
    'font.family': ['serif'],
    'font.size': 10.0,
    'grid.linestyle': '--',
    'legend.facecolor': '.9',
    'legend.frameon': True,
    'savefig.transparent': True,
    'text.color': '.0',
    'xtick.labelsize': 'small',
    'ytick.labelsize': 'small'}
```

```python
dados = pd.read_excel(r'dados_bruto\FTIR_BIM.xlsx')
x44 = dados['x44']
y44 = dados['y44']
x62 = dados['x62']
y62 = dados['y62']
```

```python
plt.style.use(['seaborn-whitegrid', 'seaborn-paper', 'seaborn-muted', config])

fig, ax = plt.subplots(1,1)
ax.tick_params(direction='in', which='both',top=True,right=True,labelsize=12)


ax.plot(x44, y44, c = 'g')
ax.set_xlabel('Número de onda ($cm^{-1}$)')
ax.set_ylabel('Transmitância (%)')
ax.set_title('Espectro FTIR da Amostra 44')
ax.invert_xaxis()

plt.savefig(r'graficos\AM_44_FTIR_BIM.png', dpi = 200)
```

```python
plt.style.use(['seaborn-whitegrid', 'seaborn-paper', 'seaborn-muted', config])

fig, ax = plt.subplots(1,1)
ax.tick_params(direction='in', which='both',top=True,right=True,labelsize=12)


ax.plot(x62, y62, c = 'r')
ax.set_xlabel('Número de onda ($cm^{-1}$)')
ax.set_ylabel('Transmitância (%)')
ax.set_title('Espectro FTIR da Amostra 62')
ax.invert_xaxis()

plt.savefig(r'graficos\AM_62_FTIR_BIM.png', dpi = 200)
```
