```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
config = {'axes.spines.right': False,
    'axes.spines.top': False,
    'axes.edgecolor': '.4',
    'axes.labelcolor': '.0',
    'axes.titlesize': 'large',
    'axes.labelsize': 'medium',
    'figure.autolayout': True,
    'figure.figsize': (4.5, 3.5),
    'font.family': ['serif'],
    'font.size': 10.0,
    'grid.linestyle': '--',
    'legend.facecolor': '.9',
    'legend.frameon': True,
    'savefig.transparent': True,
    'text.color': '.0',
    'xtick.labelsize': 'small',
    'ytick.labelsize': 'small'}
```

```python
dados = pd.read_excel(r'dados_bruto\UVVIS _BIM.xlsx')
x = dados['wavelength']
y44 = dados['AM_44_BIM']
y62 = dados['AM_62_BIM']
```

```python
plt.style.use(['seaborn-whitegrid', 'seaborn-paper', 'seaborn-muted', config])

fig, ax = plt.subplots(1,1)
ax.tick_params(direction='in', which='both',top=True,right=True,labelsize=12)


ax.plot(x, y44, c = 'g')
ax.set_xlabel('Comprimento de onda (nm)')
ax.set_ylabel('Absorb√¢ncia (u.a)')
ax.set_title('Espectroscopia UV-vis da Amostra 44')

plt.savefig(r'graficos\AM_44_UVVIS_BIM.png', dpi = 200)
```

```python
fig, ax = plt.subplots(1,1)
ax.tick_params(direction='in', which='both',top=True,right=True,labelsize=12)


ax.plot(x, y62, c = 'r')
ax.set_xlabel('Comprimento de onda (nm)')
ax.set_ylabel('Absorb√¢ncia (u.a)')
ax.set_title('Espectroscopia UV-vis da Amostra 62')

plt.savefig(r'graficos\AM_62_UVVIS_BIM.png', dpi = 200)
```
