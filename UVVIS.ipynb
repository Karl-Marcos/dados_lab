```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
config = {'axes.spines.right': False,
    'axes.spines.top': False,
    'axes.edgecolor': '.4',
    'axes.labelcolor': '.0',
    'axes.titlesize': 'large',
    'axes.labelsize': 'medium',
    'figure.autolayout': True,
    'figure.figsize': (4.5, 3.5),
    'font.family': ['serif'],
    'font.size': 10.0,
    'grid.linestyle': '--',
    'legend.facecolor': '.9',
    'legend.frameon': True,
    'savefig.transparent': True,
    'text.color': '.0',
    'xtick.labelsize': 'small',
    'ytick.labelsize': 'small'}
```

```python
dados = pd.read_excel(r'dados_bruto\UVVIS_BIM.xlsx')
x = dados['wavelength']
y44 = dados['AM_44_BIM']
y62 = dados['AM_62_BIM']
```

```python
plt.style.use(['seaborn-whitegrid', 'seaborn-paper', 'seaborn-muted', config])

fig, ax = plt.subplots(1,1)
ax.tick_params(direction='in', which='both',top=True,right=True,labelsize=12)


ax.plot(x, y44, c = 'g')
ax.set_xlabel('Comprimento de onda (nm)')
ax.set_ylabel('Absorbância (u.a)')
ax.set_title('Espectro UV-Vis da Amostra 44')

plt.savefig(r'graficos\AM_44_UVVIS_BIM.png', dpi = 200)
```

```python
fig, ax = plt.subplots(1,1)
ax.tick_params(direction='in', which='both',top=True,right=True,labelsize=12)


ax.plot(x, y62, c = 'r')
ax.set_xlabel('Comprimento de onda (nm)')
ax.set_ylabel('Absorbância (u.a)')
ax.set_title('Espectro UV-Vis da Amostra 62')

plt.savefig(r'graficos\AM_62_UVVIS_BIM.png', dpi = 200)
```

```python
def normalizar(dados):
    fator = max(dados['Absorbance'])
    for i in range(len(dados['Absorbance'])):
        dados['Absorbance'][i] = dados['Absorbance'][i]/fator
    return dados
```

```python
Amostra_2_Dialise = normalizar(pd.read_csv(r'dados_bruto/Amostra_2_Dialise.txt'))
Amostra_2_Glomalina = normalizar(pd.read_csv(r'dados_bruto/Amostra_2_Glomalina.txt'))
Amostra_2_Salting = normalizar(pd.read_csv(r'dados_bruto/Amostra_2_Salting.txt'))
Amostra_4_Dialise = normalizar(pd.read_csv(r'dados_bruto/Amostra_4_Dialise.txt'))
Amostra_4_Glomalina_2 = normalizar(pd.read_csv(r'dados_bruto/Amostra_4_Glomalina_2.txt'))
Amostra_4_Salting = normalizar(pd.read_csv(r'dados_bruto/Amostra_4_Salting.txt'))
```

```python
fig, ax = plt.subplots(1,1)
ax.tick_params(direction='in', which='both',top=True,right=True,labelsize=12)


ax.plot(Amostra_2_Glomalina['Wavelength (nm)'], Amostra_2_Glomalina['Absorbance'], label = 'Centrifugada')
ax.plot(Amostra_2_Salting['Wavelength (nm)'], Amostra_2_Salting['Absorbance'], label = 'Salting')
ax.plot(Amostra_2_Dialise['Wavelength (nm)'], Amostra_2_Dialise['Absorbance'], label = 'Dialise')
ax.set_xlabel('Comprimento de onda (nm)')
ax.set_ylabel('Absorbância (u.a)')
ax.set_title('Espectro UV-Vis da Amostra 2')
ax.legend()

plt.savefig(r'graficos\Amostra_2.png', dpi = 200)
```

```python
fig, ax = plt.subplots(1,1)
ax.tick_params(direction='in', which='both',top=True,right=True,labelsize=12)


ax.plot(Amostra_4_Glomalina_2['Wavelength (nm)'], Amostra_4_Glomalina_2['Absorbance'], label = 'Centrifugada')
ax.plot(Amostra_4_Salting['Wavelength (nm)'], Amostra_4_Salting['Absorbance'], label = 'Salting')
ax.plot(Amostra_4_Dialise['Wavelength (nm)'], Amostra_4_Dialise['Absorbance'], label = 'Dialise')
ax.set_xlabel('Comprimento de onda (nm)')
ax.set_ylabel('Absorbância (u.a)')
ax.set_title('Espectro UV-Vis da Amostra 4')
ax.legend()

plt.savefig(r'graficos\Amostra_4.png', dpi = 200)
```
